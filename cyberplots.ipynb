# Graphical features part

**Cyberpunk Style with Matplotlib**

Futuristic neon glow for data visualization.

**1 — The Basis**

Let’s make up some numbers, put them in a Pandas dataframe and plot them.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({"A": [1, 3, 9, 5, 2, 1, 1],
                   "B": [4, 5, 5, 7, 9, 8, 6]})

df.plot(marker="o",figsize=(10,8)) # marker - 0 for dots, figsize x,y
plt.show()
```

**2 — The Darkness**

Let’s customize it by using Seaborn’s dark style, as well as changing background and font colors.

```python
plt.style.use("seaborn-dark")

for param in ["figure.facecolor", "axes.facecolor", "savefig.facecolor"]:
  plt.rcParams[param] = "#212946"  # bluish dark grey

for param in ["text.color", "axes.labelcolor", "xtick.color", "ytick.color"]:
  plt.rcParams[param] = "0.9"  # very light grey 

df.plot(marker="o",figsize=(10,8))
plt.grid(color="#2A3459")
plt.show()
```

It looks more interesting now, but we need our colors to shine more against the dark background.

**3 — The Light**

```python
fig, ax = plt.subplots()

colors = [
    "#08F7FE",  # teal/cyan
    "#FE53BB",  # pink
    "#F5D300",  # yellow
    "#00ff41", # matrix green
]

df.plot(marker="o", ax=ax, color=colors, figsize=(10,8))
plt.grid(color="#2A3459")
plt.show()
```

**4 — The Glow**

To get the neon look/ To make it shine, we redraw the lines multiple times, with low alpha value and slighty increasing linewidth. The overlap creates the glow effect.

```python
fig, ax = plt.subplots()
n_lines = 10
diff_linewidth = 0.05
alpha_value = 0.3

for n in range(1, n_lines+1):
    df.plot(marker="o",
            linewidth=2+(diff_linewidth*n),
            alpha=alpha_value,
            legend=False,
            ax=ax,
            color=colors,
            figsize=(10,8))
    
plt.grid(color="#2A3459")    
plt.show()
```

**5 — The Finish**

For some more fine tuning, we can color the area below the line and adjust the axis limit.

```python
# Color the areas below the lines:
fig, ax = plt.subplots()

for column, color in zip(df, colors):
    ax.fill_between(x=df.index,
                    y1=df[column].values,
                    y2=[0] * len(df),
                    color=color,
                    alpha=0.1)

df.plot(marker="o", ax=ax, color=colors, figsize=(10,8)) 
ax.grid(color="#2A3459")

plt.show()
```
